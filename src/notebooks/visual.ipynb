## READ ENTITY VECTOR 

```python
# read_entity_embedding.py
import numpy as np
from tqdm import tqdm

ENTITY_PATH = "/home/xuanhau/CodeSpace/freelancer/infor_recommendation/MINDlarge_train/MINDlarge_train/entity_embedding.vec"

print("Đang đọc entity_embedding.vec...")
entity2vec = {}
with open(ENTITY_PATH, 'r', encoding='utf-8') as f:
    for line in tqdm(f, total=232554):
        values = line.rstrip().split()
        if len(values) != 101: 
            continue
        entity_id = values[0]
        vector = np.asarray(values[1:], dtype=np.float32)
        entity2vec[entity_id] = vector

print(f"Đã load {len(entity2vec):,} thực thể")
# Test: Apple Inc. → Q312
print("Apple Inc. (Q312):", entity2vec.get("Q312", "Không có")[:5])
```
